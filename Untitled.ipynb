```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

# --- 1. VERİ YÜKLEME ---
try:
    df = pd.read_csv('netflix_titles.csv')
except:
    url = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-04-20/netflix_titles.csv'
    df = pd.read_csv(url)

# --- 2. TEMİZLEME ---
df_clean = df.dropna(subset=['title', 'release_year', 'listed_in', 'type']).copy()
df_clean['main_genre'] = df_clean['listed_in'].apply(lambda x: x.split(',')[0])

# --- 3. VERİ ÖN İZLEME ---
print("\nVERİ ÖN İZLEME:")
display(df_clean.head())

# --- 4. İÇERİK TÜRÜ SEÇİMİ ---
print("\n1. Film | 2. Dizi")
type_input = input("Seçiminiz (1/2): ")

selected_type = "Movie" if type_input == "1" else "TV Show"

# --- 5. TÜRLERİ LİSTELE ---
type_filtered = df_clean[df_clean['type'] == selected_type].copy()
genres = sorted(type_filtered['main_genre'].unique())

print(f"\n{selected_type} için türler:")
for i, g in enumerate(genres[:10], 1):
    print(f"{i}. {g}")

# --- 6. KULLANICI SEÇİMİ ---
genre_input = int(input("\nTür numarası gir: "))

if genre_input < 1 or genre_input > len(genres[:10]):
    print("Geçersiz seçim!")
    exit()

selected_genre = genres[genre_input - 1]

print("\nSeçilenler:")
print("Tür:", selected_type)
print("Kategori:", selected_genre)

# --- 7. FİLTRELEME ---
final_df = type_filtered[
    type_filtered['main_genre'].str.contains(selected_genre, case=False)
].copy()

# --- 8. SIRALAMA ---
final_df = final_df.sort_values(by='release_year', ascending=False)

# --- 9. ÖNERİLER ---
top_content = final_df[['title', 'release_year']].head(5)

print("\nONERILEN ICERIKLER:")
display(top_content)

# --- 10. EN İYİ ÖNERİ ---
if not top_content.empty:
    best = top_content.iloc[0]

    print("\nEN IYI ONERI")
    print("Isim:", best['title'])
    print("Yil:", best['release_year'])
else:
    print("Uygun içerik bulunamadı.")

# --- 11. GRAFİK ---
plt.figure(figsize=(10,5))
final_df['release_year'].value_counts().sort_index().tail(10).plot(kind='line', marker='o')
plt.title(f"{selected_genre} - Son Yıllar")
plt.xlabel("Yıl")
plt.ylabel("İçerik Sayısı")
plt.show()
```
