```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. PROBLEM VE VERİ ---
# Problem: Tür bazlı içerik analizi
# Kaynak: Kaggle Netflix Dataset

try:
    # Veriyi yükle
    df = pd.read_csv('netflix_titles.csv')
except:
    # Alternatif kaynak
    df = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-04-20/netflix_titles.csv')

# --- 2. ANALİZ VE ÖN İŞLEME ---

# Tanıtım
print("--- Veri Bilgisi ---")
print(df.info())

# Eksikler
print("\n--- Eksik Değerler ---")
print(df.isnull().sum())

# Temizleme
df_clean = df.dropna(subset=['title', 'release_year', 'listed_in', 'type']).copy()

# Ayrıştırma
df_clean['main_genre'] = df_clean['listed_in'].apply(lambda x: x.split(',')[0])
valid_genres = sorted(df_clean['main_genre'].unique()[:15])

# --- 3. ARAYÜZ VE GRAFİK ---

print("\nTürler:")
for i, genre in enumerate(valid_genres, 1):
    print(f"{i}. {genre}")

try:
    # Seçim
    choice = int(input("\nTür no: "))
    selected_genre = valid_genres[choice - 1]
    print(f"\nSeçilen: {selected_genre}")

    # Filtre
    filtered = df_clean[df_clean['main_genre'] == selected_genre].copy()

    if filtered.empty:
        print("İçerik yok.")
    else:
        # Görselleştirme
        plt.figure(figsize=(12, 6))
        year_counts = filtered['release_year'].value_counts().sort_index().tail(15)

        sns.lineplot(x=year_counts.index, y=year_counts.values, marker='o', color='red', linewidth=2.5)

        plt.title(f'{selected_genre} Artış Grafiği')
        plt.xlabel('Yıl')
        plt.ylabel('Sayı')
        plt.grid(True, linestyle='--', alpha=0.6)
        plt.show()

        # Listeleme
        print(f"\nEn yeni 5 {selected_genre}:")
        display(filtered[['title', 'release_year', 'country', 'rating']].sort_values(by='release_year', ascending=False).head(5))

except:
    print("Hatalı giriş!")
```
